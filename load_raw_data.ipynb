### Convert tables to csv for easier import to postgres

```python
import pandas as pd
```

This table includes data on google_businesses from google maps in california, the json format and unnecessary large data makes this hard to load. We will remove unnecessary columns an duplicated rows for easier loading. 

```python
google_path = "C:/Users/cbrun/OneDrive/Documents/Inspection_Project/meta-California.json"
df_g = pd.read_json(google_path,lines=True)
df_g.head(1)
```

```python
drop_columns = ['hours','state','relative_results','url','description','price','latitude','longitude']
df_g = df_g.drop(drop_columns, axis=1)
```

We can load our preview easier now with the large columns dropped

```python
print(len(df_g))
df_g.head(10)
```

drop duplicated values based on gmap_id since this is our unique identifier

```python
print(len(df_g[df_g.duplicated(['gmap_id'])]))
df_g.drop_duplicates(subset='gmap_id',inplace=True)
```

```python
#save to csv 
google_output = "C:/Users/cbrun/OneDrive/Documents/Inspection_Project/google_business.csv"
df_g.to_csv(google_output,index=False)
```
